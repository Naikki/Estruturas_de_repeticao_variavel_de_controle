Esse código escreve a tabuada de multiplicação completa dos números de 1 a 10, utilizando a estrutura repetição "for".

```python
#Tabuada de Multiplicação de 1 a 10

def tabuada_multiplicacao():
    for num1 in range(1, 11): # Multiplicador principal
        print(f"--- Tabuada do {num1} ---")
        for num2 in range(1, 11): # Multiplicador secundário
            resultado = num1 * num2
            print(f"{num1} x {num2} = {resultado}")
        print()

tabuada_multiplicacao()
```

O código calcula a soma da série alternada
S = 1/1 - 2/4 + ... ± 10/100 utilizando
um loop que itera de 1 a 10,
aplicando a fórmula (-1)^(n+1)*(1/n)
para cada termo e acumulando os resultados.

A série S = 1/1 - 2/4 + 3/9 - 4/16 + ... + 10/100 segue o padrão:
- Numerador (n): números inteiros de 1 a 10.
- Denominador: quadrado do numerador (n²).
- Sinal alternado: termos ímpares (+, n=1,3,5...) positivos, termos pares (-, n=2,4,6...) negativos.

n=1: termo = +1.0000 | soma parcial = 1.0000
n=2: termo = -0.5000 | soma parcial = 0.5000
n=3: termo = +0.3333 | soma parcial = 0.8333
n=4: termo = -0.2500 | soma parcial = 0.5833
n=5: termo = +0.2000 | soma parcial = 0.7833
n=6: termo = -0.1667 | soma parcial = 0.6167
n=7: termo = +0.1429 | soma parcial = 0.7595
n=8: termo = -0.1250 | soma parcial = 0.6345
n=9: termo = +0.1111 | soma parcial = 0.7456
n=10: termo = -0.1000 | soma parcial = 0.6456

Valor final de S: 0.645635

Verificação com a forma original (n/n²):
S = 0.645635.

```python
soma = 0.0

for n in range(1, 11):
    termo = (-1) ** (n + 1) * (1 / n)
    soma += termo
    print(f"n={n}: termo = {termo:+.4f} | soma parcial = {soma:.4f}")

print(f"\nValor final de S: {soma:.6f}")

# Verificação explícita da série original
print("\nVerificação com a forma original (n/n²):")
soma2 = 0.0
for n in range(1, 11):
    termo = n / (n ** 2)
    if n % 2 == 0:  # termos pares são negativos
        termo *= -1
    soma2 += termo
print(f"S = {soma2:.6f}")
```

### Verificador de Números Primos

Este código solicita ao usuário que insira um número inteiro maior que 1. Em seguida, ele verifica se o número fornecido é primo ou não. Um número primo é um número natural maior que 1 que não possui divisores positivos além de 1 e de si mesmo.

**Como funciona:**
1. O programa pede ao usuário para digitar um número.
2. Ele verifica se o número é menor ou igual a 1. Se for, ele solicita um número válido.
3. Para números maiores que 1, ele itera de 2 até a raiz quadrada do número.
4. Se encontrar qualquer divisor nesse intervalo, o número não é primo.
5. Se nenhum divisor for encontrado, o número é primo.

```python
def verificar_numero_primo():
    while True:
        try:
            numero = int(input("Digite um número inteiro maior que 1 para verificar se é primo: "))
            if numero <= 1:
                print("Por favor, digite um número inteiro maior que 1.")
            else:
                break
        except ValueError:
            print("Por favor, digite um número inteiro.")

    if numero == 2:
        print(f"O número {numero} é primo.")
        return

    if numero % 2 == 0:
        print(f"O número {numero} não é primo (é divisível por 2).")
        return

    e_primo = True
    for i in range(3, int(numero**0.5) + 1, 2): # Verifica apenas divisores ímpares
        if numero % i == 0:
            e_primo = False
            break

    if e_primo:
        print(f"O número {numero} é primo.")
    else:
        print(f"O número {numero} não é primo.")

# Chamada da função
verificar_numero_primo()
```

# Análise de Dados Antropométricos

Este código processa um conjunto de dados de 50 pessoas contendo:
- Altura (em metros)
- Sexo (M para masculino, F para feminino)

## Funcionalidades implementadas:
1. **Leitura dos dados**: Solicita altura e sexo de 50 pessoas com validação de entrada
2. **Cálculos estatísticos**:
   - Maior e menor altura do grupo (independente do sexo)
   - Média de altura apenas das mulheres
   - Contagem de homens e cálculo da diferença percentual entre homens e mulheres
3. **Saída formatada**: Apresenta os resultados de forma organizada

## Estrutura do código:
- Utilização de listas para armazenamento temporário
- Validação de entrada para sexo (apenas M/F) e altura (valores positivos)
- Cálculo de porcentagem com tratamento para divisão por zero
- Formatação de saída com casas decimais apropriadas

```python
def main():
    # Inicialização de variáveis
    total_pessoas = 50
    alturas = []
    sexos = []

    # Contadores específicos
    mulheres = 0
    soma_altura_mulheres = 0
    homens = 0

    print("=== CADASTRO DE DADOS ANTROPOMÉTRICOS ===")
    print(f"Por favor, insira os dados de {total_pessoas} pessoas\n")

    # Leitura dos dados de 50 pessoas
    for i in range(1, total_pessoas + 1):
        print(f"\nPessoa {i}/{total_pessoas}:")

        # Validação da altura
        while True:
            try:
                altura = float(input("Altura (em metros, ex: 1.75): "))
                if altura > 0:
                    alturas.append(altura)
                    break
                else:
                    print("Erro: A altura deve ser um valor positivo.")
            except ValueError:
                print("Erro: Digite um número válido para a altura.")

        # Validação do sexo
        while True:
            sexo = input("Sexo (M para masculino e F para feminino): ").upper()
            if sexo in ['M', 'F']:
                sexos.append(sexo)
                if sexo == 'F':
                    mulheres += 1
                    soma_altura_mulheres += altura
                else:
                    homens += 1
                break
            else:
                print("Erro: Digite apenas 'M' ou 'F'.")

    # Cálculos estatísticos
    # a. Maior e menor altura
    maior_altura = max(alturas)
    menor_altura = min(alturas)

    # b. Média de altura das mulheres
    media_mulheres = 0
    if mulheres > 0:
        media_mulheres = soma_altura_mulheres / mulheres

    # c. Diferença percentual entre homens e mulheres
    diferenca_percentual = 0
    if mulheres > 0:
        diferenca_percentual = ((homens - mulheres) / mulheres) * 100
    elif homens > 0:
        diferenca_percentual = 100  # Se não há mulheres, homens são 100% a mais

    # Apresentação dos resultados
    print("\n" + "="*50)
    print("RESULTADOS DA ANÁLISE")
    print("="*50)

    print(f"\na) ALTURAS EXTREMAS DO GRUPO:")
    print(f"   • Maior altura: {maior_altura:.2f}m")
    print(f"   • Menor altura: {menor_altura:.2f}m")

    print(f"\nb) ESTATÍSTICAS DAS MULHERES:")
    print(f"   • Quantidade de mulheres: {mulheres}")
    if mulheres > 0:
        print(f"   • Média de altura das mulheres: {media_mulheres:.2f}m")
    else:
        print(f"   • Não há mulheres no grupo para calcular a média")

    print(f"\nc) DISTRIBUIÇÃO POR SEXO:")
    print(f"   • Número de homens: {homens}")
    print(f"   • Número de mulheres: {mulheres}")
    print(f"   • Total de pessoas: {total_pessoas}")

    print(f"\n   DIFERENÇA PERCENTUAL:")
    if diferenca_percentual >= 0:
        print(f"   • Há {abs(diferenca_percentual):.1f}% mais homens que mulheres")
    else:
        print(f"   • Há {abs(diferenca_percentual):.1f}% mais mulheres que homens")

    print("\n" + "="*50)

if __name__ == "__main__":
    main()
```

# Cálculo de Imposto de Renda para Múltiplos Contribuintes

Este código realiza o cálculo do Imposto de Renda para 10 contribuintes, considerando:

## Dados de entrada para cada contribuinte:
1. **CPF** - Número do CPF do contribuinte
2. **Número de dependentes** - Para cálculo de dedução
3. **Renda mensal** - Renda bruta do contribuinte
4. **Valor do salário mínimo atual** - Fornecido uma única vez para todos os cálculos

## Lógica de cálculo implementada:
1. **Dedução por dependente**: 5% do salário mínimo por cada dependente
2. **Cálculo da renda líquida**: Renda mensal - (dedução por dependentes)
3. **Aplicação da tabela progressiva**:
   - Até 2 SM: Isento
   - 2 a 3 SM: 5% de imposto
   - 3 a 5 SM: 10% de imposto  
   - 5 a 7 SM: 15% de imposto
   - Acima de 7 SM: 20% de imposto

## Saída do programa:
Para cada contribuinte, exibe:
- CPF
- Renda bruta
- Valor deduzido por dependentes
- Renda líquida
- Alíquota aplicada
- Valor do imposto a pagar

```python
def converter_para_float(valor_str):
    """
    Converte uma string numérica no formato brasileiro para float.
    Aceita: 1.621,00 ou 1621,00 ou 1621.00 ou 1621
    """
    # Remove espaços extras
    valor_str = valor_str.strip()

    # Remove o símbolo de moeda se presente
    valor_str = valor_str.replace('R$', '').replace('$', '').strip()

    # Substitui vírgula por ponto se for o separador decimal
    if ',' in valor_str and '.' in valor_str:
        # Formato 1.621,00 -> remove pontos de milhar, substitui vírgula por ponto
        valor_str = valor_str.replace('.', '').replace(',', '.')
    elif ',' in valor_str:
        # Formato 1621,00 -> substitui vírgula por ponto
        valor_str = valor_str.replace(',', '.')

    return float(valor_str)

def main():
    # Configurações iniciais
    NUM_CONTRIBUINTES = 10

    # Entrada do salário mínimo com tratamento robusto
    while True:
        try:
            salario_str = input("Digite o valor atual do salário mínimo (ex: 1.621,00 ou 1621.00): R$ ")
            SALARIO_MINIMO = converter_para_float(salario_str)
            if SALARIO_MINIMO <= 0:
                print("Erro: O salário deve ser maior que zero.")
            else:
                break
        except ValueError:
            print("Erro: Digite um valor numérico válido.")
        except Exception as e:
            print(f"Erro: {e}. Digite um valor válido.")

    print("\n" + "="*60)
    print("SISTEMA DE CÁLCULO DE IMPOSTO DE RENDA")
    print("="*60)
    print(f"Salário mínimo informado: R$ {SALARIO_MINIMO:,.2f}".replace(',', 'X').replace('.', ',').replace('X', '.'))
    print(f"Serão processados {NUM_CONTRIBUINTES} contribuintes")
    print("="*60 + "\n")

    # Tabela de alíquotas (em salários mínimos)
    TABELA_IMPOSTO = [
        (0, 2, 0.00),      # Até 2 SM - Isento
        (2, 3, 0.05),      # 2 a 3 SM - 5%
        (3, 5, 0.10),      # 3 a 5 SM - 10%
        (5, 7, 0.15),      # 5 a 7 SM - 15%
        (7, float('inf'), 0.20)  # Acima de 7 SM - 20%
    ]

    # Processamento dos contribuintes
    for i in range(1, NUM_CONTRIBUINTES + 1):
        print(f"\n{'='*50}")
        print(f"CONTRIBUINTE {i}/{NUM_CONTRIBUINTES}")
        print('='*50)

        # Entrada de dados do contribuinte
        cpf = input("CPF (apenas números): ").strip()

        # Validação do número de dependentes
        while True:
            try:
                dependentes = int(input("Número de dependentes: "))
                if dependentes < 0:
                    print("Erro: O número de dependentes não pode ser negativo.")
                else:
                    break
            except ValueError:
                print("Erro: Digite um número inteiro válido.")

        # Validação da renda mensal
        while True:
            try:
                renda_str = input("Renda mensal (R$): ")
                renda_mensal = converter_para_float(renda_str)
                if renda_mensal < 0:
                    print("Erro: A renda não pode ser negativa.")
                else:
                    break
            except ValueError:
                print("Erro: Digite um valor numérico válido.")

        # Cálculos
        deducao_dependentes = dependentes * (0.05 * SALARIO_MINIMO)
        renda_liquida = renda_mensal - deducao_dependentes

        # Se a renda líquida for negativa, considerar como zero
        if renda_liquida < 0:
            renda_liquida = 0

        # Calcular renda líquida em salários mínimos
        renda_liquida_sm = renda_liquida / SALARIO_MINIMO

        # Determinar alíquota com base na tabela
        aliquota = 0.0
        faixa_descricao = ""

        for faixa_min, faixa_max, taxa in TABELA_IMPOSTO:
            if faixa_min <= renda_liquida_sm < faixa_max:
                aliquota = taxa
                if faixa_max == float('inf'):
                    faixa_descricao = f"Acima de {faixa_min} SM"
                else:
                    faixa_descricao = f"De {faixa_min} a {faixa_max} SM"
                break

        # Calcular imposto
        imposto = renda_liquida * aliquota

        # Função para formatar valores em reais no formato brasileiro
        def formatar_reais(valor):
            return f"R$ {valor:,.2f}".replace(',', 'X').replace('.', ',').replace('X', '.')

        # Exibir resultados
        print("\n" + "-"*50)
        print("RESULTADO DO CÁLCULO:")
        print("-"*50)
        print(f"CPF: {cpf}")
        print(f"Renda bruta: {formatar_reais(renda_mensal)}")
        print(f"Dependentes: {dependentes} (dedução: {formatar_reais(deducao_dependentes)})")
        print(f"Renda líquida: {formatar_reais(renda_liquida)}")
        print(f"Renda líquida em SM: {renda_liquida_sm:.2f}")
        print(f"Faixa de renda: {faixa_descricao}")
        print(f"Alíquota aplicada: {aliquota*100:.0f}%")

        if imposto > 0:
            print(f"IMPOSTO A PAGAR: {formatar_reais(imposto)}")
        else:
            print("IMPOSTO A PAGAR: ISENTO")

        print("-"*50)

    print("\n" + "="*60)
    print("PROCESSAMENTO CONCLUÍDO!")
    print(f"Total de contribuintes processados: {NUM_CONTRIBUINTES}")
    print("="*60)

if __name__ == "__main__":
    main()
```
